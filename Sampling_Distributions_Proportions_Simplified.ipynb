<h4>Import Python Libraries</h4>

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random as rd
import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore",DeprecationWarning)
```

<h4>Create the Simulated the Sampling Distribution</h4>

```python
#Edit the number of pennies, nickels, dimes, and quarters to create the population of coins. It is originally set as a left-skewed distribution of 1000 coins.
p = 1000
n = 1000
d = 1500
q = 6500

#Edit the sample size to see its effect on the simulated sampling distribution.
#Edit the number of samples used to simulate the sampling distribution. The Law of Large Numbers will lead the characteristics of the simulated sampling distribution to approach the characteristics of the actual sampling distribution as the number of samples increases.
samp_size = 1500
num_samples = 20000

#**************************************************
#****** DO NOT EDIT THE CODE BELOW THIS LINE ******
#**************************************************

#Create Simulated Sampling Distribution
N = p + n + d + q
prop = p/N
PHATS = [0]*num_samples
sum = 0
x = 0
for i in range(0,num_samples):
    for j in range(0,samp_size):
        x = rd.randint(0,N - 1)
        if x < p:
            sum = sum + 1
    PHATS[i] = sum/samp_size
    sum = 0
phats = pd.DataFrame(PHATS)

#Graph Simulated Sampling Distribution
f = plt.figure()
f, axes = plt.subplots(nrows = 1, ncols = 1,  figsize=(8,4), sharex=False, sharey = True)

axes.hist(phats, ec = 'b', bins=15)
axes.set_xlabel('Sample Proportions', labelpad = 5)
axes.set_ylabel('Frequency', labelpad = 5)
axes.set_title('Simulated Sampling Distribution of phat')

#Print details of population and simulated sampling distribution.
print('')
print('********************************************************************************************************************')
print('The proportion of pennies in your ',N,' coins is ', format(prop,'.4f'),'.')
print('********************************************************************************************************************')
print('The mean value of your simulated sampling distribution for sample proportions from samples of ',samp_size,' coins is ', format(phats[0].mean(),'.4f'), ' and the standard deviation of the sampling distribution of sample proportions is ', format(phats[0].std(),'.4f'),'.')
print('********************************************************************************************************************')
print('')
```
