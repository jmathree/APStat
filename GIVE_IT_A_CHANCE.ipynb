```python
#Simulation for Topic 4.2 "Give It a Chance!"
#Set n as number of flips (100 in AP Daily Video), s for size of streak (8 in video), t for number of trials in next cell. 
#Then from menu above click "Run" and "Run all cells"
#Law of Large Numbers says as t approaches infinity, simulated probability should approach theoretical probability.
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

If you want to look at the impact changing parameters of the simulation, adjust the code in the next cell, then run the Notebook again.

```python
# Change n for number of flips, change s for size of streak, t for number of trials of n flips
n = 100
s = 8
t = 50
#Don't change anything below this cell unless you know how to code!
```

```python

<h2 style="color:red">DO NOT CHANGE THE CODE BELOW THIS CELL UNLESS YOU WANT TO COMPLETELY CHANGE THE SIMULATION!!!</h2></style>
```

```python
def coin_flips(n):
    streak = 0
    flip = [0]*n # 1 is heads, 0 is tails
    curr_streak = 0 #current streak in iteration
    streak = 0 #longest streak in flip
    for i in range(0,n):
        flip[i] = rnd.randint(0,1)
        if i > 0:
            if flip[i] == flip[i - 1]:
                curr_streak = curr_streak + 1
            else:
                curr_streak = 1
            if curr_streak > streak:
                streak = curr_streak
    return(streak)
    
```

```python
streaks = [0]*t
count = 0
for k in range(0,t):
    streaks[k] = coin_flips(n)
    if streaks[k] > s - 1:
        count = count + 1
FLIPS = pd.DataFrame(streaks)
#FLIPS = FLIPS.value_counts().rename_axis('LONGEST_STREAK').reset_index(name='FREQUENCY')
print('The simulated probability of a streak of {} or more heads or {} or more tails in {} flips is {}.'.format(s,s,n,count/t))
```

```python
print('The simulated probability of a streak of {} or more heads or {} or more tails is {}.'.format(s,s,count/t))
FLIPS[0].plot.hist(grid=True, bins=(FLIPS[0].max()-FLIPS[0].min()+2), range=(FLIPS[0].min()-1,FLIPS[0].max()+1), rwidth=1.0,color='b',edgecolor='k')
plt.title('Topic 4.2 Daily Video: Give it a Chance!')
plt.xlabel('Longest Streak')
plt.ylabel('Frequency')
```
