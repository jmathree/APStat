<h1>Simulation for Topic 4.2 AP Daily Video #2</h1>

AP Daily instructor Josh Sawyer references a simulation in his 2nd AP Daily Video for Topic 4.2. He calls the simulation "Sharpshooter" The code in this Notebook runs that simulation.

The code is preset for the parameters in Mr. Sawyer's video example. The shooter is an 82% free throw shooter. The record for made free throws before the first miss is 16. Number of trials is 200.

```python
#Simulation for Topic 4.2 "Sharpshooter"
#Set n as number of flips (100 in AP Daily Video), s for size of streak (8 in video), t for number of trials in next cell. 
#Then from menu above click "Run" and "Run all cells"
#Law of Large Numbers says as t approaches infinity, simulated probability should approach theoretical probability.
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

```python
# Change pct for free throw percentage
# Change rec for record number of shots to be broken.
# Change t for number of trials
pct = 0.82 
rec = 16
t = 200
#Don't change anything below this cell unless you know how to code!
```

<h2 style="color:red;">DO NOT CHANGE THE CODE BELOW THIS CELL UNLESS YOU WANT TO COMPLETELY CHANGE THE SIMULATION!!!</h2></style>

```python
def free_throws():
    streak = 0 # number of shots made
    made = True
    while made == True:
        shot = rnd.randint(1,100)
        if shot < 100*pct + 1: 
            streak = streak + 1
        else:
            made = False
    return(streak)
```

```python
num_made = [0]*t
count = 0
for k in range(0,t):
    num_made[k] = free_throws()
    if num_made[k] > rec:
        count = count + 1
SHOTS_MADE = pd.DataFrame(num_made)
print('The simulated probability of breaking the record of {} shots made before the first miss is {}.'.format(rec,count/t))
```

```python

SHOTS_MADE[0].plot.hist(grid=True, bins=(SHOTS_MADE[0].max()-SHOTS_MADE[0].min()+2), range=(SHOTS_MADE[0].min()-1,SHOTS_MADE[0].max()+1), rwidth=1.0,color='b',edgecolor='k')
plt.title('Topic 4.2 Daily Video: Sharpshooter')
plt.xlabel('Free Throws Made Before First Miss')
plt.ylabel('Frequency')
```
