<h1>Exploring Sampling Distributions</h1>

Introduction to Sampling Distributions for AP Statistics.

<h3>Definition of a Sampling Distribution</h3>

<ul><li>For a Sample Mean -- The distribution of ALL sample means for every possible sample of size n, with replacement, for a particular population.</li>
<li>For a Sample Proportion -- The distribution of ALL sample proportions for every possible sample of size n, with replacement, for a particular population</li></ul>


Think about a population of 100 coins--pennies, nickels, dimes, and quarters. If we select every possible random sample, with replacement, of 5 coins from the population, assuming the probability of selecting each individual coin is the same, and calculate the mean value of the 5 coins for every possible sample, the distribution of ALL those sample means is the sampling distribution of the sample mean for 5 randomly selected coins from the population. 

Let's start with a simple example population that we can easily calculate the population mean and standard deviation of. 25 pennies, 25 nickels, 25 dimes, and 25 quarters.

<h3>Import Python Libraries</h3>

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

<h3>Create Coin Population</h3>


```python
p = 25
n = 25
d = 25
q = 25
N = p + n + d + q
COINS = N*[0]

for i in range(0,p):
    COINS[i] = 0.01
for i in range(p,p+n):
    COINS[i] = 0.05
for i in range(p+n,p+n+d):
    COINS[i] = 0.10
for i in range(p+n+d,N):
    COINS[i] = 0.25

coins = pd.DataFrame(COINS)
coins.describe()
print('The mean value of your ',N,' coins is $', format(coins[0].mean(),'.4f'), ' and the standard deviation of the values of your coins is $', format(coins[0].std(),'.4f'),'.')
```

<h3>Determine Sample Size and Number of Samples in Simulated Sampling Distribution</h3>

```python
samp_size = 10
num_samples = 100
```

<h3>Generate Sample Means</h3>

```python
import random as rd
samp_means = [0]*num_samples
sum = 0
x = 0
for i in range(0,num_samples):
    for j in range(0,samp_size):
        x = rd.randint(0,N - 1)
        if x < p:
            sum = sum + 0.01
        elif x < p + n:
            sum = sum + 0.05
        elif x < p + n + d:
            sum = sum + 0.10
        else:
            sum = sum + 0.25
    samp_means[i] = sum/samp_size
    sum = 0
SAMP_MEANS = pd.DataFrame(samp_means)
print('The mean value of your simulated sampling distribution for samples of ',samp_size,' coins is $', format(SAMP_MEANS[0].mean(),'.2f'), ' and the standard deviation of the sampling distribution is $', format(SAMP_MEANS[0].std(),'.2f'),'.')
```

```python
print('Simulated Sampling Distribution for ', num_samples, ' samples of size ', samp_size, ' vs Distribution of Population.')
f = plt.figure()
f, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(8,4), sharex=False, sharey = True)

axes[0].hist(coins, ec = 'b')
axes[0].set_xlabel('Coin Values', labelpad = 5)
axes[0].set_ylabel('Frequency', labelpad = 5)
axes[0].set_title('Population')

axes[1].hist(SAMP_MEANS, ec = 'b')
axes[1].set_xlabel('Sample Mean Values', labelpad = 5)
axes[1].set_ylabel('Frequency', labelpad = 5)
axes[1].set_title('Simulated Sampling Distribution')
```
